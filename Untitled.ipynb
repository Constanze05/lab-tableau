```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv('olist_customers_dataset.csv')
data1.head(50)
```

```python
customer_dict = dict(zip(data1.customer_city, data1.customer_unique_id))
customer_dict
```

```python
data2 = pd.read_csv('olist_geolocation_dataset.csv')
```

```python
data3 = pd.read_csv('olist_sellers_dataset.csv')
data3.head(5)
```

```python
seller_dict = dict(zip(data3.seller_city, data3.seller_id))
seller_dict
```

```python
data2['costomers_id'] = data2['geolocation_city'].map(customer_dict)

data2['seller_id'] = data2['geolocation_city'].map(seller_dict)
data2.head(20)
```

```python
data2.shape
```

```python
dsnull = data2.isnull().sum()

print(dsnull)
```

```python
data2 = data2.dropna() 
```

```python
data2 = data2.drop_duplicates()
```
